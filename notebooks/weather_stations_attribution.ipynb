#### Weather station attribution

We will limit the number of weather stations to be used for each route to the up to 10 closest within a 20km radius

```python
import os
os.chdir('..')
print(os.path.abspath(os.curdir))
```

```python
from sqlalchemy.orm import Session
from willthisfreeze.dbutils import load_routes, read_config, get_engine

conf = read_config()
engine =  get_engine(conf)   
session =  Session(engine)
```

```python
#filter for routes in france only (id 14274)
routes = load_routes(session, countryId=14274)
```

Example with one route:

```python
route = routes[0]
print(route.lat, route.lon)
print(route.routeId)
```

```python
from willthisfreeze.datapreparation.weather_stations_attribution import load_stations_within_radius, filter_stations

routelat = route.lat
routelon = route.lon

stations = load_stations_within_radius(session=session, lat=routelat, lon=routelon, radius=20)

stationsId = filter_stations(route, stations)

#list of ids
print(stationsId)
```

```python
from sqlalchemy.orm import Session
from willthisfreeze.datapreparation.weather_stations_attribution import update_routes_station_mapping

update_routes_station_mapping(session=session, route=route, station_ids=stationsId)

session.close()
```
